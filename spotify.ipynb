```python
harry=data.groupby('genre')['popularity'].mean().sort_values(ascending=False)

print(harry)
plt.figure(figsize=(12,6))
harry.plot(kind='bar',color='skyblue',edgecolor='black')
plt.title('genre of music with populariy',fontsize=16)
plt.xlabel('genre of music',fontsize=12)
plt.ylabel('popularity',fontsize=12)
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
data.groupby('genre').agg({'popularity':['mean','sum','max'],
                                 "energy":['mean','max','sum']}).sort_values(('energy','mean'),ascending=False)
```

```python
harry2=data.groupby('genre')['energy'].mean().sort_values(ascending=False).head(10)
print(harry2)
plt.figure(figsize=(12,6))
harry2.plot(kind='bar',color='red',edgecolor='black')
plt.title('Energy claassifcation with genre',fontsize=18)
plt.xticks(rotation=45,ha='right')
plt.xlabel('genre of music',fontsize=12)
plt.ylabel('energy',fontsize=12)
plt.tight_layout()
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.show()
```

```python
def get_status(score):
    if (score>=80):
       return "HIGH"
    elif(score>=50 and score<=79):
       return "MEDIUM"
    else:
       return "LOW"

data['popularity_status']=data['popularity'].apply(get_status)
data
    
```
